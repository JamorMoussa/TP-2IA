# TP2 : Manipulation des images

```python
import cv2
import matplotlib.pyplot as plt
import numpy as np
```

```python
def plot_rgb(img):
    plt.imshow(cv2.cvtColor(img, cv2.COLOR_RGB2BGR))
```

#### Load image :

```python
img = cv2.imread("./src/img.jpg")
```

```python
plot_rgb(img)
```

#### Load image in gray scale :

```python
img_g = cv2.imread("./src/img.jpg", 0)
```

```python
plot_rgb(img_g)
```

#### Display image:

```python
cv2.imshow("window", img)
```

#### Write image :

```python
cv2.imwrite("./src/writed_img.png", img)
```

```python
width , height = 200, 200 
emp_img = np.empty((width, height, 3), dtype = np.uint8)
plot_rgb(emp_img)
```
