# Segmentation dâ€™une image

```python
import cv2 
import numpy as np
import matplotlib.pyplot as plt
```

```python
def cv2_imshow(img):
    plt.imshow(cv2.cvtColor(img, cv2.COLOR_RGB2BGR))
```

```python
def ax_plot(img, ax):
    ax.imshow(cv2.cvtColor(img, cv2.COLOR_RGB2BGR))
```

```python
img = cv2.imread("../img.jpg", 0)
```

```python
cv2_imshow(img)
```

## 1. Seuillage simple :

```python
types = [cv2.THRESH_BINARY, cv2.THRESH_BINARY_INV, cv2.THRESH_TRUNC, cv2.THRESH_TOZERO, cv2.THRESH_TOZERO_INV]
```

```python
def threshold_type(img):
    for i in range(5):
        ax = plt.subplot(3, 2, i+1)
        img_ = cv2.threshold(img, 135, 255, types[i])[1]
        ax_plot(img_, ax)
threshold_type(img)
```

```python
cv2_imshow(cv2.threshold(img, 135, 255, cv2.THRESH_BINARY)[1])
```

## 2. Seuillage adaptatif

```python
adpt_method = [cv2.ADAPTIVE_THRESH_MEAN_C, cv2.ADAPTIVE_THRESH_GAUSSIAN_C]
```

```python
def adpt_threshold(img, adpt_m):
    for i in range(5):
        ax = plt.subplot(3, 2, i+1)
        try:
            img_ = cv2.adaptiveThreshold( img, 255, adpt_m, types[i], 11, 2)
        except:
            pass
        ax_plot(img_, ax)
```

```python
for i in (0,1):
    adpt_threshold(img, adpt_method[i])
```

##### Comparaison :
On observe que le seuillage adaptatif donne de meilleurs rÃ©sultats que le seuillage simple, car il permet de mieux dÃ©terminer les contours.
